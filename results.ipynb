```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sea

df = pd.read_excel('results.xlsx')
df.head(12)
```

```python
dfAccuracy = df.sort_values(by='Accuracy', ascending=False)
sea.barplot(x = 'Model Name', y = 'Accuracy', data = dfAccuracy)
plt.xticks(rotation=90)
plt.show()
```

```python
dfPrecision = df.sort_values(by='Precision', ascending=False)
sea.barplot(x = 'Model Name',y = 'Precision' ,data = dfPrecision)
plt.xticks(rotation=90)
plt.show()
```

```python
dfRecall = df.sort_values(by='Recall', ascending=False)
sea.barplot(x = 'Model Name',y = 'Recall' ,data = dfRecall)
plt.xticks(rotation=90)
plt.show()
```

```python
dfF1 = df.sort_values(by='F1', ascending=False)
sea.barplot(x = 'Model Name', y = 'F1', data = dfF1)
plt.xticks(rotation=90)
plt.show()
```

```python
df['sum'] = df.loc[:,['Accuracy','Precision', 'Recall', 'F1']].sum(axis=1)
test = df.sort_values(['sum'], ascending=[False])
test
```
